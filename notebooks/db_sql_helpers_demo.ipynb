üìò Using dfs_to_sql_ctes to Query Pandas DataFrames with PostgreSQL Syntax

This cell demonstrates how to convert one or more pandas DataFrames into PostgreSQL-compatible CTEs using the helper function dfs_to_sql_ctes().
Once converted, you can write pure SQL against those DataFrames directly inside the notebook.

üîß 1. Build SQL CTEs From DataFrames

We call dfs_to_sql_ctes(cur, ‚Ä¶), passing:

‚Ä¢ a psycopg2 cursor (cur)
‚Ä¢ any number of pandas DataFrames
‚Ä¢ each DataFrame is automatically assigned a CTE name based on the notebook variable name

Example:

cte_sql = dbsh.dfs_to_sql_ctes(
    cur,
    df_roles,
    df_groups
)


This produces a WITH clause containing CTE definitions for each DataFrame.

üîç 2. Query the DataFrames Using PostgreSQL

Now that the DataFrames exist as SQL CTEs, we can query them with real SQL syntax, just like temporary tables.

Example filtering df_roles:

sample_df_query = "SELECT * FROM df_roles WHERE id = 1000 OR id = 1001;"
df_sample_sql = cte_sql + sample_df_query

üìä 3. Execute the SQL Into Pandas

We execute the full SQL string using pandas:

df_sample = pd.read_sql(df_sample_sql, conn)
display(df_sample)


This returns a pandas DataFrame containing the SQL-filtered subset of df_roles.

‚úÖ Summary

‚Ä¢ dfs_to_sql_ctes turns DataFrames into SQL CTEs using mogrify (type-safe).
‚Ä¢ You can then filter, join, and transform them using real PostgreSQL syntax.
‚Ä¢ Perfect for debugging SQL logic interactively in Jupyter notebooks.

```python
cte_sql = dbsh.dfs_to_sql_ctes(
    cur,
    df_roles,
    df_groups
)


##filter the dataframe df_roles via sql syntax (LFG) instead of pandas (EW)
sample_df_query = "SELECT * FROM df_roles where id = 1000 or id = 1001;"

df_sample_sql = cte_sql + sample_df_query

df_sample = pd.read_sql(df_sample_sql,conn)

display(df_sample)


```
